```python
#Read Source file (Output of Wav2Vec2Phoneme)
sf = pd.read_csv(current_language+"/"+current_source, sep = '\t',header=0)
sf.head()
```

```python
sf['file_name'] = sf['file_name'].str.replace('English/','')
sf = sf[['file_name', 'transcription']]
sf.head()
```

```python
#Read Target file (Transcript from Mozilla Common Voice)
tf = pd.read_csv(current_language+"/"+current_target, sep = '\t', header=0)
tf.head()
```

```python
#Merge the source and target to obtain the correct mapping between them on the basis of name of audio clip
tf_sf = tf.merge(sf, how = 'inner', left_on = 'path', right_on = 'file_name')
tf_sf.head()
```

```python
#Merge the source and target to obtain the remainig transcription as training data for SentencePiece
tf = tf.merge(sf, how = 'outer', left_on = 'path', right_on = 'file_name')
tf = tf[tf['file_name'].isna()]
tf.head()
```

```python
import sentencepiece as spm
import csv

#Function to read a specific column from a TSV file and write to a new file
def extract_column_to_file(input_file, output_file, column_index):
    with open(input_file, 'r', encoding='utf-8') as infile, \
         open(output_file, 'w', encoding='utf-8') as outfile:
        reader = csv.reader(infile, delimiter='\t')
        for row in reader:
            if len(row) > column_index:
                outfile.write(row[column_index] + '\n')

#Extract the desired column for training
extract_column_to_file(current_language+'/Output/train.tsv', current_language+'/Output/train_extracted.txt', column_index=0)  #Change column_index as needed

#Train SentencePiece model from the extracted file
spm.SentencePieceTrainer.train('--input='+current_language+'/Output/train_extracted.txt --model_prefix=m --vocab_size=6000')

#makes segmenter instance and loads the model file (m.model)
sp = spm.SentencePieceProcessor()
sp.load('m.model')

#Function to tokenize text using the SentencePiece model
def tokenize_text(text, sp_processor):
    return ' '.join(sp_processor.encode_as_pieces(text))

#Extract the desired column for testing
extract_column_to_file(current_language+'/Output/target.tsv', current_language+'/Output/target_extracted.txt', column_index=0)  #Change column_index as needed

#Open the extracted test file and an output file for the tokenized text
with open(current_language+'/Output/target_extracted.txt', 'r', encoding='utf-8') as test_file, \
     open(current_language+'/Output/tokenized_output.tsv', 'w', encoding='utf-8') as output_file:

    #Read each line in the test file, tokenize it, and write to the output file
    for line in test_file:
        tokenized_line = tokenize_text(line.strip().replace('▁',''), sp)
        output_file.write(tokenized_line + '\n')

print("Tokenization complete. Output saved to 'tokenized_output.tsv'")
```

```python
def calculate_unalignment_ratio(alignment_pairs, sentence_length, transcription_length):
    #Extracting indices from the pairs
    first_indices = {pair.split('-')[0] for pair in alignment_pairs[0]}
    second_indices = {pair.split('-')[1] for pair in alignment_pairs[0]}

    # print(len(first_indices))
    # print(len(second_indices))

    #Counting unique unaligned indices based on the ranges of sentence and transcription lengths
    unaligned_first = sentence_length - len(first_indices)
    unaligned_second = transcription_length - len(second_indices)

    #Total unaligned count
    total_unaligned = unaligned_first + unaligned_second
    # print(unaligned_first)
    # print(unaligned_second)
    # Dividing by the transcription length to find the ratio
    unalignment_ratio = (total_unaligned / sentence_length) * 100

    return unalignment_ratio
```
